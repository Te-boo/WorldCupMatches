```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/VAC dataVisualisation with python/csv files/WorldCupMatches.csv')

# Data Analysis
num_rows, num_columns = df.shape
print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_columns}')

print("Column names:")
for column in df.columns:
    print(column)

print("\nRows from index 10 to 20:")
print(df.iloc[10:21])

print("\nSummary statistics:")
print(df.describe(include='all'))

# Data Cleaning
print("\nMissing values:")
print(df.isnull().sum())

# Handling missing values (mean or median)
numeric_columns = df.select_dtypes(include=['number']).columns# Use mean to replace missing values
# Data Visualization
sns.pairplot(df)
plt.show()

sns.boxplot(data=df)
plt.show()

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
plt.show()

# Replace 'categorical_column' and 'numerical_column' with your actual column names
sns.barplot(x='Home Team Name', y='Home Team Goals', data=df)
plt.legend(title='Categories')
plt.show()

# Replace 'Half-time Home Goals' with an actual numerical column name from your dataset
df['Half-time Home Goals'].hist(bins=30) # Replaced 'numerical_column' with 'Half-time Home Goals'
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Half-time Home Goals') # Replaced 'numerical_column' with 'Half-time Home Goals'
plt.show()

sns.histplot(df['Half-time Home Goals'], kde=True) # Replaced 'numerical_column' with 'Half-time Home Goals'
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Distribution Plot of Half-time Home Goals') # Replaced 'numerical_column' with 'Half-time Home Goals'
plt.show()
# Replace 'numerical_column1' and 'numerical_column2' with actual numerical column names
sns.scatterplot(x='Half-time Home Goals', y='Half-time Away Goals', data=df) # Replaced placeholders with actual column names
plt.xlabel('Half-time Home Goals') # Replaced placeholders with actual column names
plt.ylabel('Half-time Away Goals') # Replaced placeholders with actual column names
plt.title('Scatterplot between Half-time Home Goals and Half-time Away Goals') # Replaced placeholders with actual column names
plt.show()

# Additional functions
print("\nIsnull DataFrame:")
print(df.isnull())

print("\nCount of missing values:")
print(df.isnull().sum())

 #Replace 'numerical_column' with an actual numerical column name from your dataset
print(f"\nMax value in 'Half-time Home Goals': {df['Half-time Home Goals'].max()}") # Replaced 'numerical_column' with 'Half-time Home Goals'
print(f"Min value in 'Half-time Home Goals': {df['Half-time Home Goals'].min()}") # Replaced 'numerical_column' with 'Half-time Home Goals'
```
